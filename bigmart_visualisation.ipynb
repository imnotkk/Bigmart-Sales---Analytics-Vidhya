```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("train_bigmart.csv")
```

```python
data.head()
```

```python
data.Item_Fat_Content.unique()
```

```python
data.Item_Fat_Content = data.Item_Fat_Content.map({"low fat":"Low Fat","LF":"Low Fat","reg":"Regular"})
```

```python
sns.boxplot(y="Item_Outlet_Sales",x="Item_Fat_Content",data=data)
#This shows that both contents are bought almost the same.
```

```python
sns.lmplot(y="Item_Outlet_Sales",x="Item_Weight",data=data)
```

```python
sns.lmplot(x="Item_Visibility",y="Item_Outlet_Sales",data=data,hue="Outlet_Type")
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(y="Item_Outlet_Sales",x="Item_Type",data=data)
```

```python
plt.figure(figsize=(18,10))
sns.lmplot(x="Item_MRP",y="Item_Outlet_Sales",data=data,hue="Outlet_Type")
#Items are not usually not priced at rounded prices like 200, but priced 199 thats why there are peaks and then almost absent Item_sales.
```

```python
plt.figure(figsize=(18,10))
sns.boxplot(x="Outlet_Identifier",y="Item_Outlet_Sales",data=data)
#Outlet27 has higher sales than any others while 10,19 have they least.
```
